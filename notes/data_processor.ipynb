## Preparing the graph

### load the graph

```python
fin_new = open("C:/Users/SID-DRW/OneDrive - Delft University of Technology/Desktop/new_boundaries_0204/GENERATING_GML/second_round/etreeGML.json")
j_new = json.loads(fin_new.read())
```

```python
def get_nodes_from_jsonfile(jfile):
    
    node_names = []
    nodes = []
    for spacelayer in jfile['SpaceLayers']:

        for i in jfile['SpaceLayers'][spacelayer]:
            node_names.append(i)
            for nd in node_names:
                node_id = jfile['SpaceLayers'][spacelayer][nd]['geometry']['boundaries']
            nodes.append(node_id)

    return nodes

nodes_new = get_nodes_from_jsonfile(j_new)
print(nodes_new)
```

```python
def get_edges(j):
    for sl in j['SpaceLayers']:

        edges_modified = []
        for node_key, node_value in j["SpaceLayers"][sl].items():
            source_boundary = node_value["geometry"]["boundaries"]
            for edge in node_value.get("edges", []):
                destination_node = j["SpaceLayers"][sl][edge["destination"]]
                destination_boundary = destination_node["geometry"]["boundaries"]
                edges_modified.append([source_boundary, destination_boundary])

    return edges_modified

edges_new = get_edges(j_new)
print(edges_new)
```

```python
def get_node_coords(nodes, jfile):

    node_coords = {}
    for i in nodes:
        coord = jfile['vertices'][i]
        node_coords[i] = coord

    return node_coords

node_coords_new = get_node_coords(nodes_new, j_new)
print(node_coords_new)
print(len(node_coords_new))
```

### Set coords, name and area attributes to the graph's nodes

```python
# set coords
G_new = nx.Graph()
G_new.add_nodes_from(nodes_new)
G_new.add_edges_from(edges_new)
nx.set_node_attributes(G_new, node_coords_new, 'coords')

print(G_new.nodes(data=True))
print(len(G_new))
```

```python
# zheli
def get_name_area_attributes(j_file):

    result = {}

    # Extract CellSpace information
    cellspaces = j_file['PrimalSpaceFeatures']['CellSpace']
    for i, (cell_id, cell_info) in enumerate(cellspaces.items()):
        result[i] = {
            'name': cell_info.get('name', 'NA'),
            'area': cell_info['attributes'].get('area', 'NA')
        }

    return result

name_area_attr = get_name_area_attributes(j_new)
print(name_area_attr)
```

```python
def enlarge_name_area_arrt(old_attr,G):
    default_value = {'name': 'NA', 'area': 'NA'}
    current_max_key = max(old_attr.keys())

    for i in range(1, (len(G.nodes)-len(old_attr)+1)):
        old_attr[current_max_key + i] = default_value

    return old_attr


new_arrt = enlarge_name_area_arrt(name_area_attr,G_new)
print(new_arrt)
```

```python
nx.set_node_attributes(G_new,new_arrt)
print(G_new.nodes(data=True))
print(G_new.edges(data=True))
```

### Processing patient journey data (get them ready for DES code)

```python
def get_specific_path(original_path, original_path_name):
# Dictionary to store the generated lists
    
    fixed_start = []
    # Loop through the original list and stop when you encounter an element starting with 'INTERACTIONSTATION'
    for element in original_path:
        if element.startswith('INTERACTIONSTATION'):
            break
        fixed_start.append(element)

    fixed_end = []
    # Reverse the list and loop through it
    for element in reversed(original_path):
        if element.startswith('INTERACTIONSTATION'):
            break
        fixed_end.append(element)

    # Since we iterated in reverse, reverse the collected elements to get the original order
    fixed_end.reverse()
    
    interaction_stations = [element for element in original_path if 'INTERACTIONSTATION' in element]

    generated_paths = {}

    # Loop through each interaction station and generate the corresponding list
    for station in interaction_stations:
        # Create the list by combining the fixed start, the specific interaction station, and the fixed end
        path = fixed_start + [station] + fixed_end
        # Create a name for the list

        ###list_name = f'medical_path_1_{station[-4:].lower()}'
        list_name = f'{original_path_name}_{station[-4:].lower()}'
        # Store the generated list in the dictionary
        generated_paths[list_name] = path

    return generated_paths
```

## Getting the shortest patient path according to patient journey 

Based on above patient path lists and the graph, find the shortest paths between each pair of adjacent rooms in the patient path list.

The resulting list will be the shortest patient path.

```python
def get_single_real_path(path, G):
    real_path_1 = []

    # Iterate through pairs of adjacent nodes in path_1
    for i in range(len(path) - 1):
        start = path[i]
        end = path[i + 1]
        
        # Find the shortest path between the two nodes
        shortest_path = nx.shortest_path(G, source=start, target=end)
        
        # Add the shortest path to real_path_1
        # Skip the first node of the shortest path if it's not the first pair to avoid duplication
        if i > 0:
            real_path_1.extend(shortest_path[1:])
        else:
            real_path_1.extend(shortest_path)

    return real_path_1

def get_all_real_paths(all_id_path,G):

    real_paths = []
    for path in all_id_path:
        real_path = get_single_real_path(path, G)
        real_paths.append(real_path)

    return real_paths
```

```python
def get_node_id_path(node_name_path,G):

    node_id_path = []
    for node in node_name_path:
        node_id = get_node_by_name(G,node)
        node_id_path.append(node_id)
    
    return node_id_path

def get_id_path_for_all_paths(all_paths,G):

    all_medical_Node_id_path = []
    for name,path in all_paths.items():
        meidical_node_id_path = get_node_id_path(path, G)
        all_medical_Node_id_path.append(meidical_node_id_path)

    return all_medical_Node_id_path
```

# Calculating difficulty in wayfinding

### Get random paths according to patient journey data for assessing difficulty in wayfinding
The above patient shortest paths are ideal situations in which patients find their destinations in shortest distance.

However, sometimes patient get lost when finding their destinations and take the wrong path.

These are the codes that simulate the situations when patients take the wrong path by generating random patient paths according to the original patient journey data.

Specifically, finding random path instead of shortest path between each pair of adjacent 'stations' in patient journey.

```python
import random
import networkx as nx

def get_single_real_path_random(path, G):
    real_path_1 = []

    # Helper function for random walk
    def random_walk(source, target, G):
        current_node = source
        walk_path = [current_node]

        # Continue walking until you reach the target node
        while current_node != target:
            neighbors = list(G.neighbors(current_node))
            
            # If no neighbors exist (disconnected node), break out
            if not neighbors:
                break
            
            # Randomly choose the next node
            current_node = random.choice(neighbors)
            walk_path.append(current_node)
        
        return walk_path

    for i in range(len(path) - 1):
        start = path[i]
        end = path[i + 1]

        # Perform a random walk from start to end
        random_path = random_walk(start, end, G)

        # Add the random path to real_path_1
        if i > 0:
            real_path_1.extend(random_path[1:])  # Skip first node if not first pair
        else:
            real_path_1.extend(random_path)

    return real_path_1

def get_all_real_paths_random(all_id_path,G):

    real_paths = []
    for path in all_id_path:
        real_path = get_single_real_path_random(path, G)
        real_paths.append(real_path)

    return real_paths
```
